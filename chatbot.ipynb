```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

## reteriver

```python
retervier=vector_store.as_retriever(search_kwargs={"k":3})
```

## LLm

```python
import os
os.environ["OPENAI_API_KEY"] = "################################"

from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
    model_name="mistralai/mistral-7b-instruct",  # or "deepseek/deepseek-chat"
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key=os.environ["OPENAI_API_KEY"]
)
```

## retrive chain

```python
from langchain.chains import RetrievalQA
qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retervier,
    return_source_documents=True
)
```

```python
print(qa_chain("explain about rag?")["result"])
```

## Evaluation

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from rouge_score import rouge_scorer
import nltk
nltk.download('punkt')


sim_model = SentenceTransformer('all-MiniLM-L6-v2')

def calc_semantic_similarity(text1, text2):
    emb1 = sim_model.encode([text1])
    emb2 = sim_model.encode([text2])
    return cosine_similarity(emb1, emb2)[0][0]

def calc_rouge_l(reference, prediction):
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    scores = scorer.score(reference, prediction)
    return scores['rougeL'].fmeasure

def groundedness(answer, source_docs):

    combined_sources = " ".join([doc.page_content for doc in source_docs])
    return calc_semantic_similarity(answer, combined_sources)

def context_relevance(question, source_docs):
    combined_sources = " ".join([doc.page_content for doc in source_docs])
    return calc_rouge_l(question, combined_sources)

def answer_relevance(question, answer):
    return calc_semantic_similarity(question, answer)
```

## Query the chatbot and evaluate

```python
query = input("Ask question: ")
result = qa_chain(query)

answer = result["result"]
source_docs = result["source_documents"]

print("\nAnswer:\n", answer)
print("\nSource Snippet:\n", source_docs[0].page_content[:500])

# Calculate evaluation metrics
gnd = groundedness(answer, source_docs)
ctx_rel = context_relevance(query, source_docs)
ans_rel = answer_relevance(query, answer)

print(f"\nEvaluation Metrics:")
print(f"Groundedness (answer supported by sources): {gnd:.3f}")
print(f"Context Relevance (query vs sources): {ctx_rel:.3f}")
print(f"Answer Relevance (query vs answer): {ans_rel:.3f}")
```
